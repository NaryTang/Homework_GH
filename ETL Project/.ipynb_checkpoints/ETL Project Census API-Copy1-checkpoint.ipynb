Loop through all years (2013 - 2017) to retrieve all data

```python
#create list to call range of years
year= np.arange(2013,2018,1)

#create list to store appended data
census_data_list = []
```

```python
#loop through each year and append to census data list
for y in year:
    c = Census(api_key, year=y)
    census_data_list.append(c.acs5.get(("NAME", "B09019_001E", "B09019_030E", "B09019_027E"), {'for': 'state:*'}))
```

```python
len(census_data_list)
```

```python
len(census_data_list[4])
```

```python
All_Data = pd.DataFrame(census_data_list[0] + census_data_list[1] + census_data_list[2] + census_data_list[3] + census_data_list[4])
All_Data.head()
len(All_Data)
```

```python
#create list of years to append to All_Data dataframe
year= np.arange(2013,2018,1)

list_of_years = []

startyear = 2012
for i in range (260):
    if i %52 == 0:
        startyear += 1
        list_of_years.append(startyear)
    else:
        list_of_years.append(startyear)

        

print(list_of_years)
# print(len(list_of_years))
```

```python
year = pd.Series(list_of_years)
len(year)
```

```python
All_Data["Year"] = year
```

```python
All_Data.head()
```

```python
#rename columns
New_df = All_Data.rename(columns = {
    "NAME": "State",
    "B09019_001E": "Total_Households",
    "B09019_030E": "Single_Female_Households",
    "B09019_027E": "Single_Male_Households"})
New_df.head()
```
