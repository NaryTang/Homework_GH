```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load, encoding='utf8')
ride_data_df = pd.read_csv(ride_data_to_load, encoding='utf8')

# city_data_df.head()
# ride_data_df.head()

# Combine the data into a single dataset
merge_data_df = pd.merge(ride_data_df, city_data_df, on="city")

# Display the data table for preview
merge_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
#filter by type and use grouby to extract values for "Urban"
urban_total_rides = merge_data_df[(merge_data_df["type"]=="Urban")].groupby("city")["ride_id"].count()
urban_avg_fare = merge_data_df[(merge_data_df["type"]=="Urban")].groupby("city")["fare"].mean()
urban_total_drivers = merge_data_df[(merge_data_df["type"]=="Urban")].groupby("city")["driver_count"].mean()

#filter by type and use grouby to extract values for "Suburban"
suburban_total_rides = merge_data_df[(merge_data_df["type"]=="Suburban")].groupby("city")["ride_id"].count()
suburban_avg_fare = merge_data_df[(merge_data_df["type"]=="Suburban")].groupby("city")["fare"].mean()
suburban_total_drivers = merge_data_df[(merge_data_df["type"]=="Suburban")].groupby("city")["driver_count"].mean()

#filter by type and use grouby to extract values for "Rural"
rural_total_rides = merge_data_df[(merge_data_df["type"]=="Rural")].groupby("city")["ride_id"].count()
rural_avg_fare = merge_data_df[(merge_data_df["type"]=="Rural")].groupby("city")["fare"].mean()
rural_total_drivers = merge_data_df[(merge_data_df["type"]=="Rural")].groupby("city")["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(urban_total_rides, urban_avg_fare, s=urban_total_drivers*10, facecolors="coral", edgecolors="black", alpha=1, label="Urban", linewidth=0.75)
plt.scatter(suburban_total_rides,suburban_avg_fare,s=suburban_total_drivers*10, facecolors="skyblue", edgecolors="black", alpha=1, label="Suburban", linewidth=0.75)
plt.scatter(rural_total_rides,rural_avg_fare,s=rural_total_drivers*10, facecolors="gold", edgecolors="black", alpha=1, label="Rural", linewidth=0.75)


# Incorporate the other graph properties
plt.grid(True)
plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
legend = plt.legend(title="City Types", loc="best")

# Incorporate a text label regarding circle size
legend.legendHandles[0]._sizes=[50]
legend.legendHandles[1]._sizes=[50]
legend.legendHandles[2]._sizes=[50]

# Save Figure
plt.savefig("Pyber.pdf")
plt.savefig("Pyber.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
total_fare = merge_data_df.groupby("type")["fare"].sum()
colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(total_fare, explode=explode, labels= types, colors = colors, autopct="%.1f%%", shadow=True, startangle=160)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("Pyber_Fare%.pdf")
plt.savefig("Pyber_Fare%.png")

# Show plot
plt.show()
```

## Total Rides by City Type

```python
# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
total_rides = merge_data_df.groupby("type")["ride_id"].count()
colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(total_rides, explode=explode, labels= types, colors = colors, autopct="%.1f%%", shadow=True, startangle=160)
plt.title("% of Total Rides by City Type")
plt.show()

# Save Figure
plt.savefig("Pyber_Rides%.pdf")
plt.savefig("Pyber_Rides%.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
#in this chart we use the original city data frame because the merged dataframe creates duplicates
total_drivers = city_data_df.groupby(["type"])["driver_count"].sum()

colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(total_drivers, explode=explode, labels= types, colors = colors, autopct="%.1f%%", shadow=True, startangle=160)
plt.title("% of Total Drivers by City Type")
plt.show()

# Save Figure
plt.savefig("Pyber_Drivers%.pdf")
plt.savefig("Pyber_Drivers%.png")

# Show Figure
plt.show()
```
