```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load, encoding='utf8')
ride_data_df = pd.read_csv(ride_data_to_load, encoding='utf8')

# city_data_df.head()
# ride_data_df.head()

# Combine the data into a single dataset
merge_data_df = pd.merge(ride_data_df, city_data_df, on="city")

# Display the data table for preview
merge_data_df.head()
```

## Bubble Plot of Ride Sharing Data

## Total Fares by City Type

```python
# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
total_fare = merge_data_df.groupby("type")["fare"].sum()
colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(total_fare, explode=explode, labels= types, colors = colors, autopct="%.1f%%", shadow=True, startangle=160)
plt.title("% of Total Fares by City Type")

# Show plot
plt.show()

# Save Figure
plt.savefig("Pyber_Fare%.pdf")
plt.savefig("Pyber_Fare%.png")
```

## Total Rides by City Type

```python
# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
total_rides = merge_data_df.groupby("type")["ride_id"].count()
colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(total_rides, explode=explode, labels= types, colors = colors, autopct="%.1f%%", shadow=True, startangle=160)
plt.title("% of Total Rides by City Type")
plt.show()

# Save Figure
plt.savefig("Pyber_Rides%.pdf")
plt.savefig("Pyber_Rides%.png")

# Show plot
plt.show()
```

## Total Drivers by City Type

```python
# Build Pie Chart
types = ["Rural", "Suburban", "Urban"]
#in this chart we use the original city data frame because the merged dataframe creates duplicates
total_drivers = city_data_df.groupby(["type"])["driver_count"].sum()

colors = ["gold", "skyblue", "coral"]
explode = (0, 0, 0.1)

plt.pie(total_drivers, explode=explode, labels= types, colors = colors, autopct="%.1f%%", shadow=True, startangle=160)
plt.title("% of Total Drivers by City Type")
plt.show()

# Save Figure
plt.savefig("Pyber_Drivers%.pdf")
plt.savefig("Pyber_Drivers%.png")

# Show Figure
plt.show()
```
